##### Bar Plot

- Simple Bar Chart
  - bar() function to draw bar graphs
  - The categories and their values represented by the first and second argument as arrays (or Python List)

```python
std = ["Ali","Bilal","Amjad","Usman","Javed"]
marks = [55, 95, 76, 47, 99]

# Bar Chart
plt.bar(std,marks)
plt.show()
```

- Add Labels (X-axis & Y-axis) & Title of the chart

```python
std = ["Ali","Bilal","Amjad","Usman","Javed"]
marks = [55, 95, 76, 47, 99]

# Labels & Title
plt.xlabel("Students")
plt.ylabel("Marks")
plt.title("Marks per Student")

# Bar Chart
plt.bar(std,marks)
plt.show()
```

- Change fonts Sizes of the Labels & Title

```python
std = ["Ali","Bilal","Amjad","Usman","Javed"]
marks = [55, 95, 76, 47, 99]

# Labels & Title
plt.xlabel("Students", fontsize = 12)
plt.ylabel("Marks", fontsize = 12)
plt.title("Marks per Student", fontsize = 15)

# Bar Chart
plt.bar(std,marks)
plt.show()
```

- Bar Plot with Errorbars

- Multiple Bar Chart 

```python
std = ["Ali","Bilal","Amjad","Usman","Javed"]
gpa = [2.56, 3.8, 3.2, 2.3, 3.9]
cgpa = [2.70, 3.59, 3.10, 2.60, 3.85]

# Labels & Title
plt.xlabel("Students", fontsize = 12)
plt.ylabel("GPA", fontsize = 12)
plt.title("GPA & CGPA per Student", fontsize = 15)

# Bars position adjustment
width = 0.4
std_index = np.arange(len(std))
std_index1 = [i + width for i in std_index]

# Bar Chart
plt.bar(std_index,gpa, width, color = "b", label="Marks")
plt.bar(std_index1,cgpa, width, color = "r", label="Rank")

# Labeling the x-axis and x-axis label adjustment
plt.xticks(std_index+width/2, std, rotation = 45) # rotation for x-axis label

plt.legend()
plt.show()
```

- Horizontal Bar Chart

```python
std = ["Ali","Bilal","Amjad","Usman","Javed"]
gpa = [2.56, 3.8, 3.2, 2.3, 3.9]
cgpa = [2.70, 3.59, 3.10, 2.60, 3.85]

# Labels & Title
plt.xlabel("Students", fontsize = 12)
plt.ylabel("GPA", fontsize = 12)
plt.title("GPA & CGPA per Student", fontsize = 15)

# Bars position adjustment
width = 0.4
std_index = np.arange(len(std))
std_index1 = [i + width for i in std_index]

# Bar Chart
plt.barh(std_index,gpa, width, color = "b", label="Marks")
plt.barh(std_index1,cgpa, width, color = "r", label="rank")

# Labeling the y-axis and y-axis label adjustment
plt.yticks(std_index+width/2, std) 

plt.legend()
plt.show()
```

##### Donut chart 
  - Method-1: Using single slice pie chart
  - Method-2: Using Circle() function with gca().add_artist()

```python
std = ["Ali","Bilal","Amjad","Usman","Javed"]
gpa = [2.56, 3.8, 3.2, 2.3, 3.9]

col = ['r' , 'g', 'b', 'y', 'c']

# Dount Chart (Method-1)
plt.pie(gpa, labels=std, colors=col, radius=1.3, textprops={"fontsize":12})
plt.pie([1],colors="w") # Method-1

plt.show()
```

```python
# Method-2: Using Circle() function with gca().add_artist()

std = ["Ali","Bilal","Amjad","Usman","Javed"]
gpa = [2.56, 3.8, 3.2, 2.3, 3.9]

col = ['r' , 'g', 'b', 'y', 'c']

# Dount Chart (Method-2)
plt.pie(gpa, labels=std, colors=col, radius=1.3, textprops={"fontsize":12})
cricle = plt.Circle(xy=(0,0), radius=1, facecolor="w")
plt.gca().add_artist(cricle)

plt.show()
```

##### Scatter Plot

- Scatter Plot with cmap (color Map) parameter

```python
demand = [10, 12, 8, 15, 16, 14, 15, 15, 16]
sales = [120, 130, 110, 150, 160, 145, 155, 150, 160]

# Labels & Title
plt.xlabel("Demand", fontsize = 12)
plt.ylabel("Sales", fontsize = 12)
plt.title("Scatter Plot: Demand Vs Sales", fontsize = 15)

colors = [20,30,10,50,60,45,55,50,60]
# Scatter Plot with camp parameter
plt.scatter(demand, sales, c= colors, cmap="plasma") 
plt.colorbar()

plt.show()
```

##### Bubble Plot

```python
demand = [10, 12, 8, 15, 16, 14, 15, 15, 16]
sales = [120, 130, 110, 150, 160, 145, 155, 150, 160]
price = [100, 300, 400, 500, 100, 400, 400, 250, 700]

# Labels & Title
plt.xlabel("Demand", fontsize = 12)
plt.ylabel("Sales", fontsize = 12)
plt.title("Bubble Plot: Demand Vs Sales with Price", fontsize = 15)

# Bubble Plot: price indicate the bubble size
plt.scatter(demand, sales, color="r", s=price)

plt.show()
```

##### Histogram

```python
# Generate 100 values of the salary variable that is Normally distributed with Mean=500 and SD=50 
salary = np.random.normal(500,50,(100))

# Histogram
plt.hist(salary)
plt.show()
```

- Histogram with specified bins and edgecolor parameters

```python
no_of_classes = int(1 + 3.3*np.log(100)) 
no_of_classes

# Histogram
plt.hist(salary, bins=no_of_classes, edgecolor="k")
plt.show()
```

- Cumulative Histogram 

```python
plt.hist(salary, bins=no_of_classes, edgecolor="k", cumulative=-1) #cumulative parameter
plt.show()
```

##### Box Plot (Box and Whisker's plot)
- Lower Whisker = Q1 - whis*(Q3-Q1)
- Upper Whisker = Q1 + whis*(Q3-Q1)

```python
# Box Plot
plt.boxplot(salary) # default whiskers = 1.5
plt.show()
```

```python
# Box Plot with whisker = 3
plt.boxplot(salary, whis=3) 
plt.show()
```

```python
# Box Plot with labels parameter
plt.boxplot(salary, labels=["Salary"])
plt.show()
```

```python
# Box Plot with showmean parameter
plt.boxplot(salary, labels=["Salary"], showmeans=True)
plt.show()
```

```python
# Horizontal Box Plot
plt.boxplot(salary, labels=["Salary"], showmeans=True, vert=False)
plt.show()
```

##### Multiple Box Plot

```python
# Generate 100 values of the salary variable that is Normally distributed with Mean=500 and SD=50 
Male_salary = np.random.normal(500,50,(100))
Female_salary = np.random.normal(450,60,(100))

Salary = [Male_salary, Female_salary]

# Box & Whisker's Plot
plt.boxplot(Salary, labels=["Male Salary", "Female Salary"])
plt.show()
```

##### Vilion Plot

```python
# Vilion Plot
plt.violinplot(salary, showmedians=True)

xLabel = ['Salary']
plt.xticks([1], xLabel)

plt.show()
```

```python
# Generate 100 values of the salary variable that is Normally distributed with Mean=500 and SD=50 
Male_salary = np.random.normal(500,50,(100))
Female_salary = np.random.normal(450,60,(100))

Salary = [Male_salary, Female_salary]

# Vilion Plot
plt.violinplot(Salary, showmedians=True)

G = ['Male_Salary', 'Female_Salary']
plt.xticks([1,2], G)


plt.show()
```
