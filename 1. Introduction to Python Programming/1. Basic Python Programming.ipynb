### 2- Variables in Python
- Varibles are the containers or objects  containing values/data. 

**Rules For naming variables**
- Do not use reserved keywords.
- No spaces are allowed in variable name
- Variable name start with string, or underscore.
- Variable name cannot be initiated with a number.
- Variable name short and descriptive
- Variable names are case sentive: lower case will be considered lower and higher case will considered higher

### Converting an abject type to another object type

```python
# Let's convert the integer number 6 to a string and a float
number = 6

print(str(number)) # convert into string & print
print(float(number)) # convert into float & print
# checking the data type 
print(type(number)) 
print(type(str(number)))
print(type(float(number)))
```

```python
# Let's conver the float number 3.14 to a string and an integer
number = 3.14

print(str(number)) # convert into string & print
print(int(number)) # convert into integer & print
# checking the data type 
print(type(number))
print(type(str(number)))
print(type(int(number)))
str(number)
```
