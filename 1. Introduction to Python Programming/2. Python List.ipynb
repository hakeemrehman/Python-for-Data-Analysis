### Membership operators
- These operators inclusing 'in' & 'not in' are employed to check if the certain value is available in the sequence of values and return True or False.

```python
rep_lis = [1,2,3,4]
print(2 in rep_lis)
print(2 not in rep_lis)
print(5 in rep_lis)
print(5 not in rep_lis)
```

#### Identity operators
- The operators is or is not are employed to control if the operands or objects to the left and right of these
operators are referring to a value stored in the same momory location and return True or False.

```python
a = 3.14
b = 1.618
print(a is b)
print(a is not b)
msg1= 'Hello, Python!'
msg2 = 'Hello, World!'
print(msg1 is msg2)
print(msg1 is not msg2)
lis1 = [3.14, 1.618]
lis2 = [3.14, 1.618]
print(lis1 is lis2) # returns False because lis1 is not the same object as lis2, even if they have the same content
print(lis1 is not lis2)

x = ["apple", "banana"]
y = ["apple", "banana"]
z = x

print(x is z) # returns True because z is the same object as x
print(x is y) # returns False because x is not the same object as y, even if they have the same content
print(x == y) # to demonstrate the difference betweeen "is" and "==": this comparison returns True because x is equal to y
```
