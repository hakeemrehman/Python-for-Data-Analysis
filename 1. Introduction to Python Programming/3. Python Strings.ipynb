## String Indexing
- String are just type of a list.
- but datatype of this list is string(str)
- It contains words letters, or anything else in quotes consider as string
- so it will have index:
    - index is an integer value.
    - starts from 0 to (lenght-of-list - 1)
    - each index shows a single letter or space in string.
    - index uniquely identifies each element in a string

```python
# Accessing specified position element of a string 
print("Element at index 0: ", stra[0]) # This index contains first letter (1st Position element)
print("Element at index 1: ", stra[1]) # Second letter
print("Element at index 2: ", stra[2]) # Third letter
print("Element at index 4: ", stra[4]) # This index contain space
```

```python
# Printing length of the string
len(stra)
```

### Slicing of a string

```python
# Get elements of a string by giving range
print("Elemnts from 0 index to 1 index: ",stra[0:2]) # it mean retrun elements form 0 to 1
print("Elemnts from 3 index to 6 index: ",stra[3:7])
print("Elemnts from 7 index to 21 index: ",stra[7:22])
## Note last number in range is no included
```

#### Negative indexing of a string

```python
# printing the last element of a string
print("The last element of a string is: ",stra[-1])
```

```python
# printing the first element of a string
print("The first element of a string is: ",stra[-22])
```

```python
# Accessing elements using Reverse indexing
print("Last three elements of the string: ", stra[-3:22])   
print("Last Seven elements of the string: ", stra[-7:22]) 
```

```python
stra[::2] # to select every second element in the variable 'stra'
```

```python
stra[0:12:2] # get every second element in range from index 0 to index 12
```

##### capitalize() Vs title() Method in a string

```python
# Convert the first letter of the text to capital letter 
str_a = 'i am learning the python programming' # creating a string

 # Convert the first letter of the text to capital letter
print("String with first capatil letter: ",str_a.capitalize())  

# Capatillizing first letter of the string
print("String with first capatil letter: ",str_a.title())
```

##### upper() Vs lower() Method in a string

```python
# Converting all the letter of a string to upper case
print("Upper case: ",str_a.upper())

# Converting all the letter of a string to lower case
print("Lower case: ",str_a.lower())
```

##### isupper() Vs islower() Method in a string

```python
print("PYTHON PROGRAMMING IS EASY TO LEARN: ".isupper()) # Give True
print("PYTHON PROGRAMMING IS EASY TO LEARN: ".islower()) # Give False
```

##### count() method in a string

```python
statment= "we are learning Python in jupyter notebook" # crearing a string

# counting letter in the string - count() is used which take 1 argument the elements to counts. it returns a counted value
# if element is not present in string it will return 0
print("Number of a's: ", statment.count('a'))
print("Number of p's: ", statment.count('p')) # lower case counted as different letter
print("Number of P's: ", statment.count('P')) # upper case counted as different letter
print("Number of o's: ", statment.count('o'))
print("Number of oo's: ", statment.count('oo')) # we can also count double latter
print("Number of yt: ", statment.count('yt')) # we can also count double latter
```

##### find() method in a string

```python
# Find an element in the string - find() function is used.
# It takes one argument element to be find and returns the index number of that element
print("'l' index is on: ", statment.find('l')) 
print("'oo' index is on: ", statment.find('oo')) # retrun the index of first element in double letter
print("'yt' index on: ", statment.find('yt'))
```

##### split() method in a string

```python
# Spliting string into words - split() is used
# It takes one argument on which basis string will be splitted. It can be space, :, #, comma etc
# Returns list of words which are separated

# Creating string with different separators
fav_fruit1 = "i like mango apple peach"  # space separated
fav_fruit2 = "i,like,gava,apple,peach"  # , separated
fav_fruit3 = "i:like:orange:apple:peach"  # : separated


print("Space separated: ", fav_fruit1.split())  # by default space separated
print("Comma separated: ", fav_fruit2.split(","))
print("Colon separated: ", fav_fruit3.split(":"))
```

##### strip() method in a string

```python
fav_fruit = "      I like mango apple peach          "
# removes or truncates the given characters from the beginning and the end of the original string
fav_fruit.strip() # it will remove all the strating & ending spaces from the string
```

```python
fav_fruit = "###############I like mango apple peach###"
# removes or truncates the given characters from the beginning and the end of the original string
fav_fruit.strip('#') # it will remove all the strating & ending # from the string
```

```python
fav_fruit = 'www.Pakistan-fruit.com'
fav_fruit.strip('wcom.') # it will remove from strat & end 'wcom.' from the string
```

```python
fav_fruit = "      I like mango apple peach          "
fav_fruit.lstrip() # it will remove only the left-side spaces from the string
```

```python
fav_fruit = "      I like mango apple peach          "
fav_fruit.rstrip() # it will remove only the right-side spaces from the string
```

##### isalpha(), isnumeric(), isalnum() methods in a string
- isalpha(): returns true if all are alphabets
- isnumeric(): returns true if all are numeric
- isalnum(): returns true if all are alphabets or numeric

```python
st1 = 'python'
st2 = '123'
st3 = 'python123'
print(st1.isalpha(), st2.isnumeric(), st3.isalnum())
print(st3.isalpha(), st1.isnumeric(), st2.isalnum())
```
