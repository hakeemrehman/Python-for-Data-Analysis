## Seaborn
- The Python visualization library Seaborn is based on matplotlib and provides a high-level interface for drawing attractive statistical graphics.

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Bar Charts

- How many male and female working in the company?

```python
sns.countplot(x="gender",data=employee)
```

#### Figure Aesthetics

#### 1. Seaborn Background styles
- Seaborn has five built-in themes to style its plots: darkgrid, whitegrid, dark, white, and ticks

```python
sns.set_style("whitegrid")
```

```python
sns.countplot(x="gender",data=employee)
```

```python
sns.set_style("ticks",{"xtick.major.size":10,"ytick.major.size":10})
sns.countplot(x="gender",data=employee)
```

##### Removing axes spines

```python
sns.countplot(x="gender",data=employee)
# Both the white and ticks styles can benefit from removing the top and right axes spines, which are not needed. 
# The seaborn function despine() can be called to remove them:
sns.despine()
```

```python
sns.set() # set / Reset the seaborn default background
sns.countplot(x="gender",data=employee)
```

- What is the average salary of managers, clerical staff etc. (i.e. Job Category) working in the company?

 - With standard deviation of the observations

```python
plt.subplots(figsize=(8,7))
sns.barplot(x="jobcat",y="salary",data=employee, ci="sd")
```

- What is the average (i.e. Median) salary of managers, clerical staff etc. (i.e. Job Category) working in the company?

```python
from numpy import median
plt.subplots(figsize=(8,7))
sns.barplot(x="jobcat",y="salary",data=employee,estimator=median,ci="sd")
```

- What is the average salary of managers, clerical staff etc. (i.e. Job Category) working in the company by gender? 

```python
plt.subplots(figsize=(8,7))
sns.barplot(x="jobcat",y="salary",hue="gender",data=employee,ci="sd")
```

- Show the standard error of the mean with the error bars (95% Confidence Interval)

```python
plt.subplots(figsize=(8,7))
sns.barplot(x="jobcat",y="salary",hue="gender",data=employee,ci=95)
```

- Control bar order by passing an explicit order

```python
plt.subplots(figsize=(8,7))
sns.barplot(x="jobcat",y="salary",hue="gender",data=employee,ci=95,order=["Clerical", "Custodial","Manager"])
```

- Add â€œcapsâ€ to the error bars:

```python
plt.subplots(figsize=(8,7))
sns.barplot(x="jobcat",y="salary",hue="gender",data=employee,ci=95,order=["Clerical", "Custodial","Manager"],capsize=.2)
```

#### Use catplot() to combine a barplot() and a FacetGrid
- This allows grouping within additional categorical variables.

```python
sns.catplot(x="educ", y="salary", hue="gender", col="jobcat", data=employee, kind="bar");
```

```python
sns.catplot(x="educ", y="salary", col="jobcat", row = "gender",hue="gender", data=employee, kind="bar");
```

```python
sns.catplot(x="educ", y="salary", hue="gender", col="jobcat", data=employee, kind="bar", height=5, aspect=.7);
```

#### Correlation & HeatMap

```python
employee.corr()
```

```python
employee_corr = employee.corr()

# irrelevant fields
fields = ['id', 'jobtime']

# drop rows
employee_corr.drop(fields, inplace=True)

# drop cols
employee_corr.drop(fields, axis=1, inplace=True)

employee_corr
```

#### Heat Map

```python
import numpy as np
# mask
mask = np.triu(np.ones_like(employee_corr, dtype=np.bool))
mask
```

```python
# adjust mask and employee_corr
mask = mask[1:, :-1]
corr = employee_corr.iloc[1:,:-1]
```

```python
mask
```

```python
corr
```

```python
# plot heatmap
sns.heatmap(corr, mask=mask, cmap='Blues',annot=True)
```

```python
sns.heatmap(corr, mask=mask, annot=True, cmap='Blues',vmin=-1, vmax=1)
```
